```python
def leave_one_out(data, yvar='AAPL'):
    data = data.copy()
    y = data.pop(yvar)
    result = {}
    for var in data.columns:
        X = data.drop([var], axis=1)
        model = ols(y=y, x=X)
        result['ex-%s' % var] = model.beta

    return result
```

```python
def winsorize(data, std_level=3, window=250, min_periods=20):
    result = data.copy()
    std = rolling_std(rets, window, min_periods=min_periods)
    cap_level = std_level * np.sign(data) * std
    result[np.abs(data) > std_level * std] = cap_level
    return result

winz = winsorize(rets)
winz_model = ols(y=winz['AAPL'], x=winz.ix[:, ['MSFT']],
                 window=250)
model.beta['MSFT'].plot(label="With outliers")
winz_model.beta['MSFT'].plot(label="Winsorized")
plt.legend(loc='best')
```

```python
def rolling_betas(returns, benchmark, window=250):
    # SPX betas
    betas = {}
    for col, ts in returns.iteritems():
        betas[col] = _rolling_beta(ts, spx, window=250)
    return DataFrame(betas)

def _rolling_beta(returns, benchmark, window=250):
    model = ols(y=returns, x={'bmk' : benchmark}, window=window,
                min_periods=100)
    return model.beta['bmk']

def get_monthly_returns(daily_returns):
    daily_index = to_index(daily_returns)
    monthly_index = daily_index.asfreq('EOM', method='pad')
    return to_returns(monthly_index)

def to_index(returns):
    # TODO: take care at beginning (start at one in prior period)
    return (1 + returns).cumprod()

def to_returns(prices):
    return prices / prices.shift(1) - 1

rets = get_monthly_returns(to_returns(close_px))

# rets = close_px / close_px.shift(1) - 1
spx = rets.pop('SPX')

betas = rolling_betas(rets, spx, window=36)
```

Application: Minimum tracking error portfolio to S&P 500
========================================================

Check out an equal-weighted portfolio for kicks
-----------------------------------------------

```python
# Some helper functions
def plot_returns(port_returns, bmk_returns):
    plt.figure()
    cum_port = ((1 + port_returns).cumprod() - 1)
    cum_bmk = ((1 + bmk_returns).cumprod() - 1)
    # cum_port = port_returns.cumsum()
    # cum_bmk = bmk_returns.cumsum()
    
    cum_port.plot(label='Portfolio returns')
    cum_bmk.plot(label='Benchmark')
    plt.title('Portfolio performance')
    plt.legend(loc='best')

def calc_te(weights, univ_rets, track_rets):
    port_rets = (univ_rets * weights).sum(1)
    return (port_rets - track_rets).std()
# port_rets = (ports.shift(1) * returns).sum(1)
```

```python
# only run the analysis since 2000
px = close_px.ix[datetime(2000, 1, 1):]
rets = px / px.shift(1) - 1

# rets = rets.drop(['AAPL'], axis=1)

annualizer = np.sqrt(250)
bmk = rets.pop('SPX') # benchmark

rets
```

```python
tickers = rets.columns
weights = Series(1. / len(tickers), index=tickers)
weights
```

```python
port_returns = (rets * weights).sum(1)
plot_returns(port_returns, bmk)
```

```python
# annualized TE
calc_te(weights, rets, bmk) * annualizer
```

**Now let's form a portfolio minimizing full sample TE**

```python
# if you have CVXOPT available
def get_opt_holdings_qp(univ_rets, track_rets):
    from cvxopt import matrix
    from cvxopt.solvers import qp

    # set up the QP for CVXOPT
    # .5 x' P x + q'x
    # P = 2 * R'R
    # q = - 2 * bmk'R

    R = univ_rets.values
    b = track_rets.values
    P = matrix(2 * np.dot(R.T, R))
    q = matrix(-2 * np.dot(R.T, b))
    result = qp(P, q)

    if result['status'] != 'optimal':
        raise Exception('optimum not reached by QP')

    return Series(np.array(result['x']).ravel(),
                  index=univ_rets.columns)

# if you don't this is a brute force solver
def get_opt_holdings_opt(univ_rets, track_rets):
    import scipy.optimize as opt
    K = len(univ_rets.columns)
    init_weights = np.ones(K) / K
    result = opt.fmin_l_bfgs_b(calc_te, init_weights,
                               args=(univ_rets.values,
                                     track_rets.values),
                               approx_grad=True, factr=1e7, pgtol=1e-7)
    return Series(result[0], index=univ_rets.columns)

clean_rets = rets.fillna(0)
clean_bmk = bmk.fillna(0)

opt_weights = get_opt_holdings_opt(clean_rets, clean_bmk)
opt_weights
```

```python
timeit opt_weights = get_opt_holdings_qp(clean_rets, clean_bmk)
```

```python
# You hope this is less
calc_te(opt_weights, rets, bmk) * annualizer
```

**Suppose we wanted a dynamic portfolio minimizing trailing 2 year TE**

```python
def rolling_min_te_portfolio(univ_returns, track_returns, window=250,
                             min_periods=100):
    ports = {}
    for i, cur_date in enumerate(univ_returns.index):
        if i < min_periods:
            continue

        if i % 100 == 0:
            print >> sys.stdout, '.',

        univ_slice = univ_returns[max(0, i - window + 1) : i]
        track_slice = track_returns[max(0, i - window + 1) : i]

        univ_slice = univ_slice.ix[:, univ_slice.count() > 0]

        univ_slice = univ_slice.fillna(0)
        track_slice = track_slice.fillna(0)
        ports[cur_date] = get_opt_holdings_qp(univ_slice, track_slice)

    return DataFrame(ports).T

ports = rolling_min_te_portfolio(rets, bmk)
```
